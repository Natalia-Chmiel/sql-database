W ramach projektu stworzyłyśmy bazę danych dla biura podróży zajmującą się wycieczkami po Europie. Projekt zrealizowałyśmy przy użyciu następujących technologii.

<a id="Spis użytych technologii"></a>
__<span style="font-size:24px;">Spis użytych technologii</span>__<br><br>

Do zarządzania bazą danych wykorzystaliśmy MariaDB/MySQL, natomiast skrypty do obsługi bazy, generowania danych napisałyśmy w pythonie. Analizę danych napisałyśmy częściowo w pythonie, częściowo w SQL. W procesie używałyśmy kilku biblotek - mysql.connector, random, datetime, matplotlib. Wszystkie działania zintegrowłyśmy w Jupyter Notebooku.

<a id="Opis plików"></a>
__<span style="font-size:24px;">Opis plików</span>__<br><br>

Projekt podzieliłyśmy na parę plików. Plik "tworzenie_bazy" jest plikiem napisanym w SQL, który najpierw usuwa wszystkie tabelki, jeżeli takie istnieją, a następnie tworzy puste tabelki od początku (polecenie CREATE TABLE). Następnym plikiem jest skrypt w pythonie "wypełnianie_bazy", gdzie z pomocą biblioteki "mysql.connector" generujemy dane do bazy. Kolejnym plikiem jest "analiza" w pythonie, w którym analizujemy dane i pokazujemy odpowiednie wizualizacje. 

<a id="Kolejność uruchamiania plików"></a>
__<span style="font-size:24px;">Kolejność uruchamiania plików</span>__<br><br>

- uruchomić "tworzenie_bazy.sql"

<a id="Zależności funkcyjne"></a>
__<span style="font-size:24px;">Zależności funkcyjne</span>__<br><br>

Zależność funkcyjna w bazie danych oznacza, że wartość jednego atrybutu jednoznacznie określa wartość innego atrybutu w tej samej tabeli.
W naszej bazie tabelki mają pełną zależność funkcyjną. Omówimy zależności funkcyjne dla każdej naszej tabelki. W naszych tabelkach korzystałyśmy z kluczy główny (primary key) oraz kluczy zapożyczony (foreign key). Omówimy każdą tabelkę. 

__<span style="font-size:24px;">Address</span>__

- address_id (primary key), city_id (foreign key)
- łączy się z tabelką staff, customers oraz city
- zależności funkcyjne: address_id --> address, city_id, postal-code, phone

__<span style="font-size:24px;">City</span>__

- city_id (primary key), country_id (foreign key)
- poza tabelką adressses łączy się z country i offers
- zależności funkcyjne: city_id --> city, country_id

__<span style="font-size:24px;">Country</span>__

- country_id (primary key)
- łączy się tylko z tabelką city
- zależności funkcyjne: country_id --> country

__<span style="font-size:24px;">Customers</span>__

- customer_id (primary key), address_id (foreign key)
- łączy się z address, orders
- zależności funkcyjne: customer_id --> first_name, last_name, email, address_id

__<span style="font-size:24px;">Finance</span>__

- offer_id (foreign key)
- łączy się z tylko z offers
- zależności funkcyjne: offer_id --> balance

__<span style="font-size:24px;">Offers</span>__

- offer_id (primary key), city_id (foreign key)
- łączy się z tabekami finance, orders i city
- zależności funkcyjne: offer_id --> city_id, duration_length, plane_number, plane_price, attractions, attractions_price, hotel_name, hotel_price, original_price, overall_price

__<span style="font-size:24px;">Orders</span>__

- order_id (primary key), offer_id, customer_id, staff_id (foreign key)
- łączy się z offers, customers oraz staff
- zależności funkcyjne: order_id --> offer_id, customer_id, staff_id, order_date, trip_start, trip_end

__<span style="font-size:24px;">Salary</span>__

- staff_id (foreign key)
- łączy się z jedynie z staff
- zależności funkcyjne: staff_id --> salary

__<span style="font-size:24px;">Staff</span>__

- staff_id (primary key)
- łączy się z salary i orders
- zależności funkcyjne: staff_id --> first_name, last_name, email, address_id
